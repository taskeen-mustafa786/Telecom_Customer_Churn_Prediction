# Predict Future Stock Prices (Short-Term)
This notebook demonstrates how to use historical stock data to predict the next day's closing price using linear regression and random forest models.

```python
# Import required libraries
import yfinance as yf
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

%matplotlib inline
```

## 1. Load data
Select a stock (e.g., Apple) and load historical data using yfinance.

```python
# Choose ticker and period
ticker = 'AAPL'
stock = yf.Ticker(ticker)
data = stock.history(period='5y')

data.head()
```

## 2. Prepare features and target
We will use Open, High, Low, Volume to predict the next day's Close price.

```python
# shift close price to represent next day

data['NextClose'] = data['Close'].shift(-1)
data = data.dropna()

features = ['Open', 'High', 'Low', 'Volume']
X = data[features]
y = data['NextClose']

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

## 3. Train models
We'll train a Linear Regression and a Random Forest model.

```python
# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)

# Random Forest
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

## 4. Evaluate models
Compute metrics and plot actual vs predicted closing prices.

```python
# predictions
pred_lr = lr.predict(X_test)
pred_rf = rf.predict(X_test)

mse_lr = mean_squared_error(y_test, pred_lr)
r2_lr = r2_score(y_test, pred_lr)

mse_rf = mean_squared_error(y_test, pred_rf)
r2_rf = r2_score(y_test, pred_rf)

print("Linear Regression MSE", mse_lr, "R2", r2_lr)
print("Random Forest MSE", mse_rf, "R2", r2_rf)
```

```python
# plot
plt.figure(figsize=(14,6))
plt.plot(y_test.index, y_test, label='Actual')
plt.plot(y_test.index, pred_lr, label='LR Predicted')
plt.plot(y_test.index, pred_rf, label='RF Predicted')
plt.legend()
plt.title('Actual vs Predicted Close Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

## 5. Results
Comparing the models' performance shows which is better for short-term prediction.
